```python
import numpy as np
import sklearn.datasets as datasets
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from numpy.polynomial.polynomial import polyfit
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
df = pd.read_csv('temp_output.csv', sep=',', index_col=0)
```

```python
x = df['area'].to_numpy().reshape(10058, 1)
x1 = df['bedrooms'].to_numpy().reshape(10058, 1)
y = df['price'].to_numpy()
```

```python
print(x.shape)
print(x1.shape)
print(y.shape)
print(type(x))
print(type(x1))
print(type(y))
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=40, test_size=0.2)
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(x1, y, random_state=40, test_size=0.2)
```

```python
regressor = linear_model.LinearRegression()
regressor.fit(X_train, y_train)
```

```python
regressor1 = linear_model.LinearRegression()
regressor1.fit(X1_train, y1_train)
```

```python
pred = regressor.predict(X_test)
```

```python
pred1 = regressor1.predict(X1_test)
```

```python
print(' score: ', regressor.score(X_train, y_train))
print(' score: ', regressor1.score(X1_test, y1_test))
```

```python
fig = plt.figure(figsize=(15,12))
plt.scatter(df['area'], y, color='blue')
plt.plot(X_test, pred, color='red')
plt.title('Price vs Area', color='black', fontsize=12)
plt.ylabel('Price')
plt.xlabel('Area')
plt.show()
```

```python
fig = plt.figure(figsize=(15,12))
plt.scatter(df['bedrooms'], y, color='blue')
plt.plot(X1_test, pred1, color='red')
plt.title('Price vs Bedrooms', color='black', fontsize=12)
plt.ylabel('Price')
plt.xlabel('Bedrooms')
plt.show()
```
